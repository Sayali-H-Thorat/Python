**Pandas Missing Data**

---



```python
import numpy as np
import pandas as pd
```

```python
d={'A':[10,20,np.nan],'B':[50,np.nan,np.nan],'C':[11,12,13]}
df=pd.DataFrame(d)
df
```

```python
#To drop missing(nan) value from dataset It will applicable for row
df.dropna()
```

```python
#To drop missing(nan) value from dataset It will applicable for column
df.dropna(axis=1)
```

```python
#It will affect the original dataframe
#df.dropna(axis=1,inplace=True)
```

```python
df
```

```python
#Another way to handle missing(nan) values
#df.fillna(value='fill value')
#mean value will be placed at nan of column A. 10+20/2=15
df['A'].fillna(value=df['A'].mean())
```

```python
df['B'].fillna(value=df['B'].mean())
```

```python
df
```

```python
#Insert missing value without using any function
df['B'].fillna(value=10)
```

```python
df
```

```python
#backward fill. It takes the values from column after nan and that fill it inplace of nan
df.bfill(axis='columns')
```

```python
df
```

```python
#forward fill. In this case nan takes values from previous row and replace it inplace of nan
df.ffill()
```
