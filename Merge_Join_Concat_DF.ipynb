**Merge,Join,Concat DataFrames**

---



```python
import pandas as pd
#Created 1st dataframe
left={'Key':['Ko','K1','K2','K3'],
      'A':['A0','A1','A2','A3'],
      'B':['B0','B1','B2','B3']}
left=pd.DataFrame(left)
left
```

```python
#Created 2nd dataframe
right={'Key':['K0','K1','K2','K3'],
       'C':['C0','C1','C2','C3'],
       'D':['D0','D1','D2','D3']}
right=pd.DataFrame(right)
right
```

```python
#Merge two dataframes
#merge also takes three parameters as 1st frame name,2nd frame name third is a common column name
pd.merge(left,right,on='Key') #It will take data of left 1st then data of right and key will take only ones
```

```python
pd.merge(right,left,on='Key')  #It will first take 2nd dataframe data first and then take 1st dataframes data
```

**Concatenation of two DataFrames**

---



```python
df1=pd.DataFrame({'A':['A0','A1','A2','A3'],
                  'B':['B0','B1','B2','B3'],
                  'C':['C0','C1','C2','C3'],
                  'D':['D0','D1','D2','D3']},index=[1,2,3,4])
df1
```

```python
df2=pd.DataFrame({'A':['A4','A5','A6','A7'],
                  'B':['B4','B5','B6','B7'],
                  'C':['C4','C5','C6','C7'],
                  'D':['D4','D5','D6','D7']},index=[5,6,7,8])
df2
```

```python
df3=pd.DataFrame({'A':['A8','A9','A10','A11'],
                  'B':['B8','B9','B10','B11'],
                  'C':['C8','C9','C10','C11'],
                  'D':['D8','D9','D10','D11']},index=[9,10,11,12])
df3
```

```python
#Concat the data from 3 dataframes
pd.concat([df1,df2,df3])
```

```python
#Concat the data from 3 dataframes
pd.concat([df1,df3,df2])
```

```python
#Concat the data from 3 dataframes->It will concat the values in tabular format
pd.concat([df1,df2,df3],axis=1)
```

**Append**

---



```python
#it adds three dataframes together
newdf=df1.append([df2,df3])
newdf
```

**Join the DataFrames**

---



```python
d1={'Name':['John','Sam','Steve'],
    'Age':[10,20,52]}
df1=pd.DataFrame(d1)
df1
```

```python
d2={'qualified':[True,False,True]}
df2=pd.DataFrame(d2)
df2
```

```python
#joins the dataframe. 2nd dataframe will get add at the end of 1st dataframe
newdf=df1.join(df2)
newdf
```
