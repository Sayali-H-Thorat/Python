```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array(['Java','Python','Excel','Sql'])
y=np.array([100,80,150,200])
plt.bar(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array(['Java','Python','Excel','Sql'])
y=np.array([100,80,150,200])
plt.barh(x,y,color='cyan') #barh for horizontal bar
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array(['Java','Python','Excel','Sql'])
y=np.array([100,80,150,200])
plt.bar(x,y,color='cyan',width=0.2)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array(['Java','Python','Excel','Sql'])
y=np.array([100,80,150,200])
plt.barh(x,y,color='cyan',height=0.5) #barh for horizontal bar
plt.show()
```
