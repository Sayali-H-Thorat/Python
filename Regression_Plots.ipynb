**Regression Plots**

---



```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
tips=sns.load_dataset('tips')
tips.head()
```

```python
tips.tail()
```

```python
tips.info()
```

```python
tips.describe()
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips)   #lm is linear model
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',palette='coolwarm')
```

**Working with markers**

---



```python
sns.lmplot(x='total_bill',y='tip',hue='sex',data=tips,palette='rainbow',markers=['o','v'])
```

```python
sns.lmplot(x='total_bill',y='tip',hue='sex',data=tips,palette='rainbow',markers=['o','v'],scatter_kws={'s':70,'alpha':0.5})   #scatter_kws will increase the size of marker. alpha will add opacity to marker
```

**Aspect**

---



```python
sns.lmplot(x='total_bill',y='tip',hue='sex',col='day',data=tips,palette='rainbow',aspect=0.7)
```
