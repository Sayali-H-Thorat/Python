```python
import numpy as np
import matplotlib.pyplot as plt
x=np.array([80,85,90,95,100,105,110,115,120,125])
y=np.array([240,250,260,270,280,290,300,310,320,330])
plt.xlabel('Sales')    #x axis name
plt.ylabel('Product')   #y axis name
plt.plot(x,y)
plt.show()
```

```python
x=np.array([80,85,90,95,100,105,110,115,120,125])
y=np.array([240,250,260,270,280,290,300,310,320,330])
plt.plot(x,y)
plt.title('Inventory')   #Title of plot
plt.xlabel('Sales')
plt.ylabel('Product')
plt.show()
```

```python
x=np.array([80,85,90,95,100,105,110,115,120,125])
y=np.array([240,250,260,270,280,290,300,310,320,330])
font1={'family':'serif','color':'blue','size':20}   #font info
font2={'family':'serif','color':'red','size':10}   #font info
plt.title('Inventory',fontdict=font1)  #Add font info as parameter here so title will access all the properties which are mentioned in the font1.
plt.xlabel('Sales',fontdict=font2)    #Adding fontdict is mandatory
plt.ylabel('Product',fontdict=font2)
plt.plot(x,y)
plt.show()
```

```python
x=np.array([80,85,90,95,100,105,110,115,120,125])
y=np.array([240,250,260,270,280,290,300,310,320,330])
font1={'family':'serif','color':'blue','size':20}   #font info
font2={'family':'serif','color':'red','size':10}   #font info
plt.title('Inventory',fontdict=font1,loc='left')  #Add title to the left hand side
plt.xlabel('Sales',fontdict=font2)
plt.ylabel('Product',fontdict=font2)
plt.plot(x,y)
plt.show()
```

**Matplotlib grids**

```python
x=np.array([80,85,90,95,100,105,110,115,120,125])
y=np.array([240,250,260,270,280,290,300,310,320,330])
font1={'family':'serif','color':'blue','size':20}
font2={'family':'serif','color':'red','size':10}
plt.title('Inventory',fontdict=font1,loc='left')
plt.xlabel('Sales',fontdict=font2)
plt.ylabel('Product',fontdict=font2)
plt.plot(x,y)
plt.grid()  #It will add grid/boxes to the graph
plt.show()
```

```python
x=np.array([80,85,90,95,100,105,110,115,120,125])
y=np.array([240,250,260,270,280,290,300,310,320,330])
font1={'family':'serif','color':'blue','size':20}
font2={'family':'serif','color':'red','size':10}
plt.title('Inventory',fontdict=font1,loc='left')
plt.xlabel('Sales',fontdict=font2)
plt.ylabel('Product',fontdict=font2)
plt.plot(x,y)
plt.grid(axis='x')  #It will add grid/boxes vertically only
#plt.grid(axis='y')  #It will add grid/boxes horizontally only
plt.show()
```

```python
x=np.array([80,85,90,95,100,105,110,115,120,125])
y=np.array([240,250,260,270,280,290,300,310,320,330])
font1={'family':'serif','color':'blue','size':20}
font2={'family':'serif','color':'red','size':10}
plt.title('Inventory',fontdict=font1,loc='left')
plt.xlabel('Sales',fontdict=font2)
plt.ylabel('Product',fontdict=font2)
plt.plot(x,y)
plt.grid(color='magenta',ls='--',lw='0.5')
#plt.grid(axis=y)  #It will add grid/boxes horizontally only
plt.show()
```
