**Distplots**

---



**Show the distribution of a univariate set of observations**

```python
import warnings
warnings.filterwarnings('ignore')
sns.distplot(tips['total_bill'])
```

```python
sns.distplot(tips['total_bill'],kde=False)   #kde-kernel density estimation
```

```python
sns.distplot(tips['total_bill'],kde=False,bins=30) #gives histogram of exact values
```

```python
sns.distplot(tips['total_bill'],bins=30)
```

**Joint Plots**

---



Allows you to basically match up two distplots for bivariate data with your choice of what kind parameter to compare with:scatter.reg,resid,kde,hex

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='scatter')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')   #reg-regression
```

**#Pairplot**

---



**Allow us to plot pairwise relationship acreoo an entire dataframe(for numerical columns) and supports a color hue argument(for categorical column)**

```python
sns.pairplot(tips)
```

```python
sns.pairplot(tips,hue='sex',palette='coolwarm')  #hue values are always categorical
```

```python
sns.pairplot(tips,hue='day',palette='coolwarm')
```

**Rugplot**

---



**rugplot is a simple concept it marks dashes for every value on univariate data**

```python
sns.rugplot(tips['total_bill'])
```
