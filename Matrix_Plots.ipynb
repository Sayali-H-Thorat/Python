```python
pv_flights=flights.pivot_table(values='passengers',index='month',columns='year')
pv_flights
```

```python
sns.heatmap(pv_flights)
```

```python
sns.heatmap(pv_flights,cmap='rainbow',linewidth=1,linecolor='white')
```
