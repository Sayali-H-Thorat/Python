**Barplot**

---



```python
sns.barplot(x='sex',y='total_bill',data=tips)
```

```python
import numpy as np
#in x and y one must it must be categorical and numerical
sns.barplot(x='sex',y='tip',data=tips)
```

```python
sns.barplot(x='sex',y='total_bill',data=tips,hue='day',palette='Set1',)
```

**Countplot**

---



```python

sns.countplot(x='sex',data=tips)
```

**Boxplot**

---



```python
sns.boxplot(x='day',y='total_bill',data=tips)
```

```python
sns.boxplot(x='day',y='total_bill',hue='smoker',data=tips,palette='rainbow')
```

**violinplot**

---



**It takes numerical and categorical column.**

```python
sns.violinplot(x='day',y='total_bill',data=tips)
```

```python
sns.violinplot(x='day',y='total_bill',hue='sex',palette='Set1',data=tips)
```

**stripplot**

---



```python
sns.stripplot(x='day',y='total_bill',data=tips,hue='sex',palette='Set1',)
```

**Swarmplot**

---



```python
sns.swarmplot(x='day',y='total_bill',data=tips)
```

```python
sns.swarmplot(x='day',y='total_bill',hue='sex',palette='coolwarm',data=tips)
```
