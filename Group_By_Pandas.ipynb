**Group By using Pandas**

---



Group by allows us to group together rows based on a column and perform an aggregate function on them


---



```python
import numpy as np
import pandas as pd
#To ignore warnings
import warnings
warnings.filterwarnings('ignore')
```

```python
data={'Company':['Google','Google','Microsoft','Microsoft','Facebook','Facebook','Amazon'],
      'Person':['Sam','John','Tina','Riva','Sara','Diva','Siya'],
      'Sales':[200,120,340,150,350,600,854]}
df=pd.DataFrame(data)
df
```

```python
#group by It will give the address
df.groupby('Company')
```

```python
#It also give same op as above
byComp=df.groupby('Company')
byComp
```

```python
#It will group the company and give the mean of that company
#It will give ans and warning both beacuse mean is applied for numerical value not for string/object.
byComp.mean()
```

```python
#It will give sum of sales by grouping company names
byComp.sum()
```

```python
#standard deviation->
byComp.std()
```

```python
df
```

```python
#It will add the sales of given company
byComp.sum().loc['Facebook']
```

```python
#It will give sum without using third variable
df.groupby('Company').sum().loc['Facebook']
```

```python
#Gives the count of company
df.groupby('Company').count()
```

```python
#It will give the maximum sales from group of comapny
df.groupby('Company').max()
```

```python
#It will give minimum sale values by grouping company
df.groupby('Company').min()
```

```python
#Describe will give all the numerical values means it will give all the functions op in single dataframe
df.groupby('Company').describe()
```

```python
#transpoase will reverse the dataframe
df.groupby('Company').describe().transpose()
```

```python
#It will describe for one single company
df.groupby('Company').describe().transpose()['Facebook']
```
