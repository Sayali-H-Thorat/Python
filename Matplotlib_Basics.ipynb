```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.array([0,5])
y=np.array([0,200])
plt.plot(x,y)
```

```python
x=np.array([1,7])
y=np.array([2,10])
plt.plot(x,y)
plt.show()
```

```python
#Plotting without line
x=np.array([1,7])
y=np.array([2,10])
plt.plot(x,y,'o')
plt.show()
```

```python
#draw line in a diagram from position(1,3) to (2,7) then(6,2) and finally to(7,10)
x=np.array([1,2,6,7])
y=np.array([3,7,2,10])
plt.plot(x,y)
plt.show()
```

```python
#Default x points
#Default x values start from 0 and so on
y=np.array([2,8,1,10,4,7])
plt.plot(y)
plt.show()
```

```python
y=np.array([2,8,1,10,4,7])
plt.plot(y,marker='*')   # '*', 'v', '^', '+', 'P','s', 'D'
plt.show()
```

```python
y=np.array([2,8,1,10,4,7])
# * will print at intersection of x and y points and : is for dotted lines and r is for red color
#marker line style and color
plt.plot(y,'*:r')
plt.show()
```

```python
y=np.array([2,8,1,10,4,7])
plt.plot(y,'d:k')  #k is for black color
plt.show()
```

```python
y=np.array([2,8,1,10,4,7])
plt.plot(y,'d-k')  #k is for black color. - is for straight line
plt.show()
```

```python
y=np.array([2,8,1,10,4,7])
plt.plot(y,'d--k')  #k is for black color  -- is for dash pattern line
plt.show()           #line references:->  '-', ':','--', '-.',
```

```python
#color references:-> r,g,b,k,m,y,w,c  support 140 colors
#change marker size
y=np.array([2,8,1,10,4,7])
plt.plot(y,marker='*',ms=20) #ms is for marker size
plt.show()
```

```python
y=np.array([2,8,1,10,4,7])
plt.plot(y,marker='*',ms=10,mec='r')   #mec for marker edge color
plt.show()
```

```python
y=np.array([2,8,1,10,4,7])
plt.plot(y,marker='*',ms=10,mec='r',mfc='k')   #mfc for marker face color
plt.show()
```

```python
y=np.array([2,8,1,10,4,7])
plt.plot(y,marker='*',ms=10,mec='#035709',mfc='#6aa89b')   #mec for marker edge color
plt.show()
```

```python
y=np.array([2,8,1,10,4,7])
plt.plot(y,marker='*',ms=10,mec='#035709',mfc='#6aa89b',ls='--')   #ls is linestyle
plt.show()
```

```python
y=np.array([2,8,1,10,4,7])
plt.plot(y,marker='*',ms=10,mec='#035709',mfc='#6aa89b',ls='--',linewidth='3.5')   #ls is linestyle
plt.show()
```
