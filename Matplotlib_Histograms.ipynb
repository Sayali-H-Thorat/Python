```python
import matplotlib.pyplot as plt
import numpy as np
y=np.array([30,35,20,5])
plt.pie(y)
plt.show()
```

```python
y=np.array([30,35,20,15])
myLabels=['blue','orange','green','red']
plt.pie(y,labels=myLabels)
plt.show()
```

```python
y=np.array([30,35,20,15])
myLabels=['blue','orange','green','red']
plt.pie(y,labels=myLabels,startangle=90)   #Startangle is 90 so piechart will start from 90 degree angle
plt.show()
```

```python
y=np.array([35,30,20,15])
myLabels=['blue','orange','green','red']
explode1=[0.2,0,0,0]    #explode will create a gap between piechart
plt.pie(y,labels=myLabels,explode=explode1)
plt.show()
```

```python
y=np.array([35,30,20,15])
myLabels=['blue','orange','green','red']
explode1=[0.2,0,0,0]    #explode will create a gap between piechart
plt.pie(y,labels=myLabels,explode=explode1,shadow=True)  #Gives shadow to pie chart
plt.show()
```

```python
y=np.array([35,30,20,15])
myLabels=['apple','orange','banana','grapes']
color=['red','orange','yellow','green']
explode1=[0.2,0,0,0]    #explode will create a gap between piechart
plt.pie(y,labels=myLabels,explode=explode1,colors=color)  #Gives color to pie chart as mentioned in the list
plt.show()
```

```python
y=np.array([35,30,20,15])
myLabels=['apple','orange','banana','grapes']
color=['black','magenta','cyan','purple']
plt.pie(y,labels=myLabels,colors=color)
plt.legend(title='Fruits')      #It will give the info about which color is for what and gives title to it
plt.show()
```

```python
y=np.array([35,30,20,15])
myLabels=['apple','orange','banana','grapes']
plt.pie(y,labels=myLabels)
plt.legend(title='Fruits',bbox_to_anchor=(1,0,0.5,1))      #bbox_to_anchor is for stop the overlapping of legend on piechart
plt.show()
```
