```python
import numpy as np
import pandas as pd
from numpy.random import randn
```

```python
#random numbers from 0 to 100
np.random.seed(101)
```

```python
df = pd.DataFrame(randn(5,4),['A','B','C','D','E'],['a','b','c','d'])
df
```

```python
# Access the value of single column
df['a']
```

```python
#Datatype of dataframe
type(df)
```

```python
#Access column values using dot operator
df.b
```

```python
#Accessing Values of two colums
df[['b','c']]
```

```python
#Add new column to dataframe which is addition two columns before new columns
df['e']=df['c']+df['d']
df
```

```python
#Delete newly added column
df.drop('e',axis=1,inplace=True)  #while working with column must add axis =1. In case of row it will take default 0. inplace=True is to add changes to original dataframe
df
```

```python
#axis =0 is not compulsory for deleting row
#df.drop('E',axis=0)
df.drop('E',inplace=True)
```

```python
import pandas as pd
#Create dataframe using dictionary
data={'Name':['Sam','John','Tina','Kim'],
      'Age':[25,21,22,26]}
df=pd.DataFrame(data,index=['Val1','Val2','Val3','Val4'])
df
```

```python
#create dataframe using dictionary by different way
#Values must be equal if not then it will give error
df=pd.DataFrame({'Col1':[10,20,30,40],
                 'Col2':[100,200,300,400]},index=['Row1','Row2','Row3','Row4'])
df
```
