**Pivot Table in Pandas**

---



```python
data={'A':['foo','foo','bar','foo','bar','bar'],
      'B':['one','one','two','two','one','one'],
      'C':['x','y','x','y','x','y'],
      'D':[10,30,20,50,40,10]}
data
```

```python
df=pd.DataFrame(data)
df
```

```python
#pivot table has 3 parameters as values index and columns
df.pivot_table(values='D',index=['A','B'],columns=['C'])
```
