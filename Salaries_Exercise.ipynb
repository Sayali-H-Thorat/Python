```python
#Average of basepay
sal['BasePay'].mean()
```

```python
#Highest overtime pay in dataset
sal['OvertimePay'].max()
```

```python
#Find the jobtitle of perticular employee
sal[sal['EmployeeName']=='JOSEPH DRISCOLL']['JobTitle']
```

```python
#Find the money earned by particular employee
sal[sal['EmployeeName']=='JOSEPH DRISCOLL']['TotalPayBenefits']
```

```python
#Find highest totalPayBenefits
sal['TotalPayBenefits'].max()
```

```python
#Find the person who has highest salary
sal[sal['TotalPayBenefits']==sal['TotalPayBenefits'].max()]
```

```python
#Find the person who has lowest salary
sal[sal['TotalPayBenefits']==sal['TotalPayBenefits'].min()]
```

```python
#Find the average basepay of all employees per year
sal.groupby('Year').mean()['BasePay']
```

```python
#Find ho many unique job titles
sal['JobTitle'].nunique()
```

```python
#Top 5 most common jobs. Find count of jobtitle which peoples do the most
sal['JobTitle'].value_counts().head(5)
```

```python
#How many jobtitle were represented by only one person in 2013
sum(sal[sal['Year']==2013]['JobTitle'].value_counts()==1)
```

```python
#How many people have the word chief in their jobtitle
def chief_string(title):
    if 'chief' in title.lower():
        return True
    else:
        return False
sum(sal['JobTitle'].apply(lambda x: chief_string(x)))
```
